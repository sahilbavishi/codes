```python
import random
board = [
    ['.', '.', ' '], ['.', '_', '.'], [' ', '.', ' ']
]
comp = 0
user = 0
draws = 0
k = True
board1 = [
    ['1,1', '1,2', '1,3'], ['2,1', '2,2', '2,3'], ['3,1', '3,2', '3,3']
]
Xc = [' ']
Yc = [' ']
print('This is how you Enter the coordinates')
print('Y being lengthwise and X being breadth_wise')
print(board1[0])
print(board1[1])
print(board1[2])
print("Enter 'S' for single player or 'D' for double player:")
v = input()


def check_draw():
    ctr = 0
    for i in range(0, 3):
        for j in range(0, 3):
            if board[i][j] == "." or board[i][j] == "_" or board[i][j] == " ":
                ctr += 1
    if ctr > 0:
        return True
    else:
        return False


# _______________________________________________________________________________________


def check_dub(b):
    # ROWS
    checker = ['A', 'B', 'C']
    for i in range(0, 3):
        for j in range(0, 3):
            checker[j] = board[i][j]
        if checker[0] == checker[1] and checker[1] == checker[2]:
            b = False
    # diagonal1
    checker = ['A', 'B', 'C']
    for i in range(0, 3):
        for j in range(i, i + 1):
            checker[j] = board[i][j]
        if checker[0] == checker[1] and checker[1] == checker[2]:
            b = False

    # Column
    checker = ['A', 'B', 'C']
    for i in range(0, 3):
        for j in range(0, 3):
            checker[j] = board[j][i]
        if checker[0] == checker[1] and checker[1] == checker[2]:
            b = False
    # D2
    checker = ['A', 'B', 'C']
    for i in range(0, 3):
        j = 2 - i
        checker[i] = board[j][i]
        if checker[0] == checker[1] and checker[1] == checker[2]:
            b = False
            break
    return b


# ____________________________s_________________________________________________


while k:
    Xc = [' ']
    Yc = [' ']
    g = True
    d = True
    u = True
    h = True
    if v.upper() == 'S':
        while d:
            try:
                x = int(input('Enter Y:')) - 1
                y = int(input('Enter X:')) - 1
                if board[x][y] != 'X' and board[x][y] != 'O':
                    board[x][y] = 'X'
                    board2 = board
                    d = False
                else:
                    print("You chose a spot that already has a ting in it, try again ")
            except IndexError:
                print('_____________________________________________')
                print('ERROR')
                print("Choose a number between 1 and 3; try again and check the guide")
                print('_____________________________________________')
            except ValueError:
                print('_____________________________________________')
                print('ERROR')
                print("Choose a number between 1 and 3; try again and check the guide")
                print('_____________________________________________')
        k = check_dub(k)
        if not k:
            user += 1
            break
        else:
            k = check_draw()
            if not k:
                draws += 1
                break
# --------------------USER INPUT---------------------------------------------------------------------------------
        while g:
            for z in range(0, 3):
                for c in range(0, 3):
                    temp = board[z][c]
                    if board[z][c] == 'X' or board[z][c] == 'O':
                        continue
                    elif board[z][c] != 'X' or board[z][c] != 'O':
                        board[z][c] = 'O'
                        h = check_dub(h)
                        board[z][c] = temp
                        if not h:
                            Xc[0] = z
                            Yc[0] = c
                            break
                if not h:
                    break
# --------------------IF COMPUTER CAN SHOOT THEIR SHOT-----------------------------------------------------------------
            if h:
                for z in range(0, 3):
                    for c in range(0, 3):
                        temp = board[z][c]
                        if board[z][c] == 'X' or board[z][c] == 'O':
                            continue

                        elif board[z][c] != 'X' or board[z][c] != 'O':
                            board[z][c] = 'X'
                            h = check_dub(h)
                            board[z][c] = temp
                            if not h:
                                Xc[0] = z
                                Yc[0] = c
                                break

                    if not h:
                        break
# --------------------IF COMPUTER CAN BLOCK ZE USER-----------------------------------------------------------------

            if h:
                s = random.randint(0, 2)
                t = random.randint(0, 2)
                if board[s][t] != 'X' and board[s][t] != 'O':
                    board[s][t] = 'O'
                    g = False
                    print('Board after computer input')
                    print('_____________________________________________')
                    print(f'{board[0]} | {board1[0]}')
                    print(f'{board[1]} | {board1[1]}')
                    print(f'{board[2]} | {board1[2]}')
                    print('_____________________________________________')
                    break
            elif not h:
                board[Xc[0]][Yc[0]] = 'O'
                g = False
                print('Board after computer input')
                print('_____________________________________________')
                print(f'{board[0]} | {board1[0]}')
                print(f'{board[1]} | {board1[1]}')
                print(f'{board[2]} | {board1[2]}')
                print('_____________________________________________')
                break

        k = check_dub(k)
        if not k:
            comp += 1
            break
# --------------------RANDOM GENERATED SPOT BECAUSE NOTHING MUCH HAPPENING--------------------------------------------
# --------------------END OF SINGLE PLAYER----------------------------------------------------------------------------
    elif v.upper() == "D":
        while d:
            try:
                print("PLAYER ONE INPUT")
                x = int(input('Enter Y:')) - 1
                y = int(input('Enter X:')) - 1
                if board[x][y] != 'X' and board[x][y] != 'O':
                    board[x][y] = 'X'
                    d = False
                else:
                    print("You chose a spot that already has a ting in it, try again ")
            except IndexError:
                print('_____________________________________________')
                print('ERROR')
                print("Choose a number between 1 and 3; try again and check the guide")
                print('_____________________________________________')
            except ValueError:
                print('_____________________________________________')
                print('ERROR')
                print("Choose a number between 1 and 3; try again and check the guide")
                print('_____________________________________________')

        print("Board after Player 1's input")
        print('_____________________________________________')
        print(f'{board[0]} | {board1[0]}')
        print(f'{board[1]} | {board1[1]}')
        print(f'{board[2]} | {board1[2]}')
        print('_____________________________________________')
        k = check_dub(k)
        if not k:
            user += 1
            break
        else:
            k = check_draw()
            if not k:
                draws += 1
                break
        while u:
            try:
                print("PLAYER TWO INPUT")
                x = int(input('Enter Y:')) - 1
                y = int(input('Enter X:')) - 1
                if board[x][y] != 'X' and board[x][y] != 'O':
                    board[x][y] = 'O'
                    u = False
                else:
                    print("You chose a spot that already has a ting in it, try again ")
            except IndexError:
                print('_____________________________________________')
                print('ERROR')
                print("Choose a number between 1 and 3; try again and check the guide")
                print('_____________________________________________')
            except ValueError:
                print('_____________________________________________')
                print('ERROR')
                print("Choose a number between 1 and 3; try again and check the guide")
                print('_____________________________________________')

            print("Board after Player 1's input")
            print('_____________________________________________')
            print(f'{board[0]} | {board1[0]}')
            print(f'{board[1]} | {board1[1]}')
            print(f'{board[2]} | {board1[2]}')
            print('_____________________________________________')
            k = check_dub(k)
            if not k:
                comp += 1
                break
            else:
                k = check_draw()
                if not k:
                    draws += 1
                    break
    else:
        print("Learn to read señor")
        print("Try again; Enter 'S' for single player or 'D' for double player:")
        v = input().upper()

if v.upper() == 'S':
    if user == 1:
        print('USER wins')
        print(board[0])
        print(board[1])
        print(board[2])

    elif comp == 1:
        print('COMPUTER wins')

    else:
        print("Y'all drew")
        print(board[0])
        print(board[1])
        print(board[2])
elif v.upper() == 'D':
    if user == 1:
        print('Player1 wins')

    elif comp == 1:
        print('Player2 wins')

    else:
        print("Y'all drew")
```
